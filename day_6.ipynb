create a program that keeps asking the user for numbers until they press =.

the program should then output:


*   sum
*   lowest val
*   highest val
*   average val

The program should not crash if the user enters a non-numerical val.



```python
from ast import Break
nums = []
print("Welcome to stat calc")
while True:
  val = input("Enter an number(or = when done to get answers)")
  if val.isnumeric():
    nums.append(float(val))
  elif val == "=":
    print(f"The sum of {nums} is {sum(nums)}.")
    print(f"Their average is {sum(nums)/len(nums)}.")
    print(f"Their lowest val is {min(nums)} while the highest is {min(nums)}.")
    break
  else:
    print("that value was invalid. the values must be numbers")
```
