```python
val = int(input("enter a value"))
if val%2 == 0 and val%3==0:
  print("the value is divisible by both 2 and 3")
elif val%2 == 0:
  print("the value is divisible by 2")
elif val%3 == 0:
  print("the value is divisible by 3")
else:
  print("the value is not divisible by both 2 and 3")
```

```python
n = input("Enter a positive integer ")
if n.isdigit() and float(n)>=0:
  n = int(n)
  total = int((n*(n+1))/2)
  print(f"the summation of all integers from 0 to {n} is {total} ")
else:
  print("invalid value, enter a positive integer")
```
