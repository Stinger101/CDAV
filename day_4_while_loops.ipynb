# Loops
## While loop

````python
# simple program that prints hello world 5 times
counter = 5 # hold the number of iterations

while counter > 0:
  print("Hello world")
  counter = counter - 1
````

````python
counter = 0

while counter<5:
  print("hello world")
  counter = counter + 1
````

````python
counter = 6

while counter>1:
  print("hello world")
  counter = counter + 1
````

````python
counter=1
while counter<=3:
  print("***"*counter)
  counter = counter + 1
````

````python
password_incorrect = True
password = "secret"
while password_incorrect:
  user_pass = input("enter the password")
  if user_pass.lower() == password.lower():
    print("Welcome back. You are logged in")
    password_incorrect = False # control
  else:
    print("Password incorrect!!! Please try again")
````

````python
not_entered_zero = True
no_of_items = 0
total=0
while not_entered_zero:
  val = float(input("Enter a value"))
  if val == 0:
    print(total/no_of_items)
    not_entered_zero = False
  else:
    total = val + total
    no_of_items= no_of_items+1
````
