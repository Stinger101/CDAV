# Body tmp calc
Write a program that asks users for their body temperature in celcius and converts it to "frnht".

```python
b_t_c = input("What is your body temp in celc?")
f_b_t_c = float(b_t_c)
b_t_f = (f_b_t_c*1.8)+32
print("Your body temp in frnht is",b_t_f)
```
